```python
import pandas as pd
import numpy as np
import csv
from collections import defaultdict
```

```python
file_name = '../../trainedModels/predictedLabelsResNet4.csv'
positive_class_name = 'violent'
negative_class_name = 'nonviolent'
tp = []
fp = []
fn = []
tn = []
```

```python

"""
Returns a pandas dataframe with columns as 'actual' class label, 'predicted' class label 
and 'image' (the file-name of the image)
"""
def get_dataframe_of_results(file):
    
    d = defaultdict(list)
    tpc = 0
    fpc = 0
    tnc = 0
    fnc = 0
    with open(file_name) as file:
        for row in file.readlines():
            rowvals = row.split(' ')
            image, predicted_class_name = rowvals[0].strip(), rowvals[1].strip()
            
            actual_class_name = image.split('/')[-2]
            d['actual'].append(actual_class_name)
            d['predicted'].append(predicted_class_name)
            d['image'].append(image.split('/')[-1])
            if actual_class_name == 'violent' and predicted_class_name == 'violent':
                tpc += 1
            elif actual_class_name == 'nonviolent' and predicted_class_name == 'nonviolent':
                tnc += 1
            elif actual_class_name == 'violent' and predicted_class_name == 'nonviolent':
                fnc+=1
            else:
                fpc += 1
                
#             print(image, actual_class_name, predicted_class_name)
            
    return pd.DataFrame(d), tpc, tnc, fpc, fnc
```

```python
df, tpc, tnc, fpc, fnc = get_dataframe_of_results(file_name)
```
